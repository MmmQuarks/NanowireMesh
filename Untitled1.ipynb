```python
import numpy as np
import matplotlib.pyplot as plt
import NanowireMesh as nwm

g = nwm.NanowireMesh(inPickle = 'temp_graph', makeEmpty = False)
```

```python
import networkx as nx
for edge in g.edges:
    g.edges[edge]['conductance'] = 1/g.edges[edge]['resistance']
    
ec = nx.eigenvector_centrality(g, weight = 'conductance')
for edge in g.edges:
    g.edges[edge]['ec'] = np.mean(
        [ec[n] for n in edge]
    )
```

```python
import pandas as pd
df = pd.DataFrame(
    [dict(I = g.edges[e]['current'], EC = g.edges[e]['ec']) for e in g.edges]
)
# normalizing
for col in df.columns:
    df[col] /= df[col].max()
```

```python
fig, ax = plt.subplots()
ax.scatter(df.I, df.EC)
plt.show()
```
